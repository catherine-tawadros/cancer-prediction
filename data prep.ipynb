```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("salaries.csv")
data.head()
```

```python
data.describe()
```

```python
def all_value_counts(column):
    counts = data[column].value_counts()
    for i in range(len(counts.index)):
        print(counts.index[i], counts[i])
all_value_counts("experience_level")
```

```python
def get_experience_level(level):
    if level == "EN":
        return 1
    if level == "MI":
        return 2
    if level == "SE":
        return 3
    if level == "EX":
        return 4
data["ex_level"] = data.experience_level.apply(get_experience_level)
```

```python
def is_manager(job):
    names = ["Manager", "Director", "Head", "Lead", "Principal"]
    for name in names:
        if name in job:
            return True
    return False
data["manager"] = data.job_title.apply(is_manager)
```
